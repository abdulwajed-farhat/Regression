```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
cardf=pd.read_csv('car-mpg.csv')
```

```python
cardf.shape
```

```python
cardf.head(20)
```

```python
cardf=cardf.drop('car_name',axis=1)
```

```python
cardf
```

```python
cardf['origin']=cardf['origin'].replace({1:'america',2:'europe',3:'asia'})
```

```python
cardf
```

```python
cardf=pd.get_dummies(cardf,columns=['origin'])
```

```python
hpisdigit=pd.DataFrame(cardf.hp.str.isdigit())
cardf[hpisdigit['hp']==False]
```

```python
cardf=cardf.replace('?',np.nan)
cardf[hpisdigit['hp']==False]
```

```python
cardf['hp'].isnull().sum()
```

```python
m=cardf['hp'].median()
m
```

```python
cardf['hp']=cardf['hp'].fillna(m)
```

```python
cardf['hp']=cardf['hp'].astype('float64')
```

```python
x=cardf.drop(['mpg'],axis=1)
y=cardf[['mpg']]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
reg_model=LinearRegression()
reg_model.fit(x_train,y_train)
```

```python
y_pred=reg_model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
acc=mean_squared_error(y_test,y_pred)
```

```python
acc
```

```python
reg_model.score(x_test,y_test)
```
